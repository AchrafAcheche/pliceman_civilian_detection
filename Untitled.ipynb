```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import uuid   # Unique identifier
import os
import time
```

```python
res = os.listdir("data/civillian")
res
for i in range(len(res)):
    os.rename('data/civillian/'+res[i], 'data/images/civilian'+'.'+str(uuid.uuid1())+'.jpg')
```

```python
!cd yolov5 && python train.py --img 320 --batch 16 --epochs 200 --data dataset.yml --weights yolov5s.pt --workers 2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp/weights/last.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
img =os.path.join('policeman.jpg')
result=model(img)
result.print()
%matplotlib inline 
plt.imshow(np.squeeze(result.render()))
plt.show()
```
